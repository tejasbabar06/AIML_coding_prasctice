```python
# Operations on 1 D array
array = np.array([10, 20, 30, 40, 50])

print(f"Array = {array}")
print(f"Array Dimensions = {array.ndim}")
print(f"Elements present in array are {array.size}")
print(f"Data Type of each element in array is {array.dtype}")
print(f"Size of each element {array.itemsize}")
print(f"Size of whole array {array.nbytes}")
print(f"Shape of array {array.shape}")
```

## Arithmetic Broadcast operations 

```python
a1 = np.array([10, 20, 30, 40 ,50])
a2 = np.array([20, 40, 60, 80, 100])
```

```python
print(f"Adding 50 to whole A1 values { a1 + 50}")
print(f"Sub 50 to whole A1 values { a1 - 50}")
print(f"Multiply 50 to whole A1 values { a1 * 50}")
print(f"Divide 50 to whole A1 values { a1 / 50}")
```

```python
print(f"Adding a2 to whole A1 values { a1 + a2}")
print(f"Sub a2 to whole A1 values { a1 - a2}")
print(f"Multiply a2 to whole A1 values { a1 * a2}")
print(f"Divide a2 to whole A1 values { a1 / a2}")
```

## Comparsion broadcast operation

```python
print(f"a1 > 30 {a1 > 30}")
print(f"a1 >= 30 {a1 >= 30}")
print(f"a1 <= 30 {a1 <= 30}")
print(f"a1 == 30 {a1 == 30}")
print(f"a1 != 30 {a1 != 30}")
print(f"a1 < 30 {a1 < 30}")
```
