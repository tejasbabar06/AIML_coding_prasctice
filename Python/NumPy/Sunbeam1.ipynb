# Statistical Measures using NumPy

```python
import numpy as np
```

```python
#10 = lower bond 
#50 = upper bond
#15 = No of values in array
# Create an array of 15 values between value 10 to 50
#data = np.random.randint(10, 50, 15)

data = np.array([31, 44, 20, 35, 21, 28, 19, 31, 40, 26, 38, 30, 18, 48, 43, 443])
print(data)
```

## Measures of central tendancy

```python
mean = np.mean(data)
print(f"Mean = {mean}")
```

```python
median = np.median(data)
print(f"Median = {median}")
```

```python
import scipy

mode = scipy.stats.mode(data)
print(f"Mode = {mode}")
```

### Measures of dispersion

```python
range = np.max(data) - np.min(data)
print(f"Range = {range}")
```

```python
#Cofficient of range

c_o_r = (np.max(data) - np.min(data)) / (np.max(data) + np.min(data))
print(f"Cofficient of range = {c_o_r}")
```

```python
# To find all quartiles

Q0 = np.quantile(data , 0)
Q1 = np.quantile(data, 0.25)
Q2 = np.quantile(data, 0.50)
Q3 = np.quantile(data, 0.75)
Q4 = np.quantile(data, 1.0)

print(f"Q0 = {Q0}, Q1 = {Q1}, Q2 = {Q2}, Q3 = {Q3}, Q4 = {Q4}")
```

```python
# Find IOR

iqr = Q3-Q1
print(f"IQR is {iqr}")
```

```python
# To find outliers

lower_bound_outliers = Q1 - (1.5 * iqr)
upper_bound_outliers = Q3 + (1.5 * iqr)
print(f"Outliers above {lower_bound_outliers} - {upper_bound_outliers}")
```

```python
 # To find outliers 

outliers = [item for item in data if(item < lower_bound_outliers) or (item > upper_bound_outliers) ]
print(f"Outliers = {outliers}")
```

```python
# To find standard devation

std = np.std(data)
print(f"Standard Devation {std}")
```

```python
# To find variance

var = np.var(data)
print(f"Variance is {var}")
```

## Measures of shpae

```python
import scipy

skewness = scipy.stats.skew(data)
print(f"Skewness is {skewness}")
```

```python
kurtosis = scipy.stats.kurtosis(data)
print(f"kurtosis is {kurtosis}")
```
