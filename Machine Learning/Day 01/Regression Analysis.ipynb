### Import required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the required dataset

```python
df = pd.read_csv('./Salary_data.csv')
```

## Visualize the relationship between X and Y

```python
plt.scatter(df['YearsExperience'], df['Salary'], label="Salary")
plt.xlabel('Experience in years')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.legend()
plt.grid()
```

```python
# Conclusion 
# - From visualization, there is a linear relationship between X and Y
```

```python
# Correlation analysis
# Correlation is relationship between two variables
# The vaue stays between -1 to 1
# -> Interpretation
#    - -1 : perfect negative relationship
#    -  0 : no relationship
#    - +1 : perfect positive relationship
df.corr()
```

### Split the data

```python
# Split data into X and Y
# It is mandatory that, X must be 2 dimensional array or data frame
# drop() will remove and entity
# axis = 0 --> Remove row with index 
# axis = 1 --> Remove column with the name

x = df.drop('Salary', axis=1)

# it is mandatory that, the y must be one dimesional array or series object
y = df['Salary']
```

### Train the model

```python
from sklearn.linear_model import LinearRegression

# Create an object of LinearRegression
model = LinearRegression()

# Train the model using x and y
model.fit(x, y)
```

### Result visualization

```python
# Plot original values 
plt.scatter(df['YearsExperience'], df['Salary'], label="Salary")

# predict salaries of all x
y_pred = model.predict(x)

# plot a regression line
plt.scatter(df['YearsExperience'], y_pred, color="red")
plt.plot(df['YearsExperience'], y_pred, color="red")

plt.xlabel('Experience in years')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.legend()
plt.grid()
```

### Interpret the model

```python
# Get the cofficients
model.coef_
```

```python
# Get the intercept
model.intercept_
```

```python
# build final formula
# m --> slope/movement along axis , c --> constant/ Intercept, x --> independent varibale
# y = mx + c
# Model formula y = 10010.1995 * Years of experience + 23487.78839975789
```
