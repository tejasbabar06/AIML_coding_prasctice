# Linear Regression

## Problem Statement
- Predict salary based on employee position

### Import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the data

```python
df = pd.read_csv('./Position_Salaries.csv')
```

### Exploratry data analysis

```python
# Get the general infromation
df.info()
```

```python
# Since the position and label have similar information let's drop position
df.drop('Position', axis=1, inplace=True)
```

```python
# Correlation analysis
df.corr()
```

```python
# Visualize the relationship
sns.pairplot(df)
```

### Split the data

```python
# Split data into x and y
x = df.drop('Salary', axis=1)
y = df['Salary']
```

### Train the model

```python
from sklearn.linear_model import LinearRegression

# Create a model
model = LinearRegression()

# Train the model
model.fit(x, y)
```

### Evaluate the model

```python
from sklearn.metrics import r2_score

# Predict the values for x
y_pred = model.predict(x)

# Get r2 Score
r2 = r2_score(y, y_pred)
print(f"R2 Score is {r2:.2f}")
```

### Result visulization

```python
# Plot the data
plt.scatter(x['Level'], y, label="Salary")

#Plot the regression line
plt.scatter(x['Level'], y_pred, color="red")
plt.plot(x['Level'], y_pred, color="red", label="Regression Line")

plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Polynomial regression")
plt.grid()
plt.legend()
```
