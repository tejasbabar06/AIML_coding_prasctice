# Binomial Distribution 

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

### Prepare the experiment

```python
# Variable for whom probablities are calculated
x = np.arange(21)

# Probablity of success
p = 0.6

# No. of trials
n = 20
```

```python
# Generate the binomial distribution 
distribution = binom.pmf(x , n, p)
distribution
```

```python
plt.plot(x, distribution)
plt.grid()
```
