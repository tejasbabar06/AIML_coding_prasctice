# Normal Distribution

### Import required libiraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Create a vaibale x
x = np.linspace(-4, 4, 100)
print(x)
```

```python
# Create a normal distribution for x
distribution = norm.pdf(x, 0, 1)
distribution
```

```python
# Plot the graph
plt.plot(x, distribution)
plt.grid()
```
